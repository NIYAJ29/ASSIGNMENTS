Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a popular data visualization library in Python used to create high-quality, publication-ready charts, graphs, and plots. It provides a variety of tools and functions for visualizing different types of data in a clear and concise manner.

Matplotlib is used for a wide range of applications, including:

Exploratory data analysis: Matplotlib can be used to quickly visualize data and gain insights into the underlying patterns and relationships.
Presentation graphics: Matplotlib provides tools for creating professional-looking charts and graphs that can be used in presentations, reports, and publications.
Web application development: Matplotlib can be used to generate dynamic visualizations in web applications.
The Pyplot module of Matplotlib provides a simple and convenient interface for creating a variety of plots. Here are five common types of plots that can be created using the Pyplot module:

1. Line plot: A line plot is a basic plot type that displays data as a series of points connected by straight lines. This plot is useful for showing trends in data over time or across different categories.

2. Scatter plot: A scatter plot is a plot type that displays data as a collection of points, where the position of each point represents its values for two different variables. This plot is useful for visualizing the relationship between two variables.

3. Bar chart: A bar chart is a plot type that displays data as a series of rectangular bars, where the height of each bar represents the values of a variable. This plot is useful for comparing values across different categories.

4. Histogram: A histogram is a plot type that displays the distribution of a continuous variable. It groups the data into intervals, or bins, and displays the frequency of data points in each bin.

5. Pie chart: A pie chart is a plot type that displays data as a circle divided into sectors, where the size of each sector represents the proportion of the data that belongs to a particular category. This plot is useful for displaying categorical data.


Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
#Note: Also add title, xlabel, and ylabel to the plot.
```

A scatter plot is a plot type used to display data as a collection of points, where each point represents the values of two different variables. One variable is plotted on the x-axis and the other on the y-axis. Scatter plots are used to explore the relationship between two variables and to identify patterns and trends in the data.

```python
import matplotlib.pyplot  as plt

plt.scatter(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('X vs Y')
```

The subplot() function is used in Matplotlib to create multiple subplots within a single figure. It is a useful function when you want to display multiple plots side by side or stacked on top of each other.

The subplot() function takes three arguments: nrows, ncols, and index. The nrows and ncols arguments specify the number of rows and columns in the subplot grid, respectively. The index argument specifies the position of the subplot within the grid.

```python
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])

# creat the sub plot
fig,axs = plt.subplots(2,2,figsize= (8,8))
axs[0,0].plot(x1,y1)
axs[0,0].set_title('line1')
axs[0,1].plot(x2,y2)
axs[0,1].set_title('line2')
axs[1,0].plot(x3,y3)
axs[1,0].set_title('line3')
axs[1,1].plot(x4,y4)
axs[1,1].set_title('line4')

fig.suptitle('four line plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

A bar plot is a type of plot in which rectangular bars are used to represent data. The length of each bar is proportional to the value of the data it represents. Bar plots are commonly used to compare the values of different groups or categories.

Bar plots are used to show the comparison between different categories or groups of data. They can be used to compare different data points or to show changes in data over time. Bar plots are especially useful when the data being compared is discrete, such as categories or groups.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit)
plt.xlabel('company')
plt.ylabel('profit')
plt.title('bar plot')
```

A box plot, also known as a box and whisker plot, is a graphical representation of the distribution of data based on five summary statistics: the minimum value, the first quartile (Q1), the median, the third quartile (Q3), and the maximum value. The box in the plot represents the interquartile range (IQR), which is the range of values between Q1 and Q3. The whiskers extend from the box to the minimum and maximum values that fall within 1.5 times the IQR.

Box plots are used to display the distribution of a dataset, identify outliers, and compare multiple datasets. They are particularly useful when comparing the distribution of multiple datasets, as they provide a clear visual representation of the differences in the data.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1,box2]


plt.boxplot(data)
                        
```
