Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a micro web framework written in Python that is used to build web applications. It is a lightweight framework that is easy to use and learn, making it a popular choice for building small to medium-sized web applications.

Some advantages of Flask Framework are:

Lightweight: Flask is a lightweight framework, which means it does not require a lot of resources to run. This makes it a great choice for building small to medium-sized applications.

Easy to learn: Flask has a simple and intuitive API that is easy to learn for beginners. It has a small core and uses extensions to add additional functionality.

Flexible: Flask is a flexible framework that can be customized to fit the needs of a project. It does not impose any particular directory structure or require the use of a specific database, making it easy to integrate with existing applications.

Large Community: Flask has a large and active community that provides support, documentation, and a wide range of extensions and plugins to enhance its functionality.

Built-in Development Server: Flask comes with a built-in development server that allows developers to quickly test and debug their applications.

Jinja2 Template Engine: Flask uses the Jinja2 template engine, which provides powerful template inheritance, filters, and macros to simplify the process of building HTML templates.

RESTful Request Dispatching: Flask provides built-in support for RESTful request dispatching, making it easy to build RESTful APIs.

Q3. What is App routing in Flask? Why do we use app routes?

In Flask, app routing is the process of mapping URLs to Python functions that handle requests for those URLs. This is done using the @app.route() decorator, which is used to associate a URL endpoint with a function that should be called when that endpoint is requested.

We use app routes in Flask to define the different endpoints of our web application and to determine which function should be executed when a user requests a particular URL. App routing allows us to organize our code and make it easy to add new endpoints or modify existing ones. It also makes it easy to create RESTful APIs by defining routes that correspond to different HTTP methods (e.g. GET, POST, PUT, DELETE).

```python
from flask import Flask
from flask import request

app = Flask(__name__)

@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

@app.route('/')
def details():
    return '''Company Name: ABC Corporation
              Location: India
              Contact Detail: 999-999-9999'''


if __name__=="__main__":
   app.run(host="0.0.0.0") 
```

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

The url_for() function is used in Flask to build URLs for specific routes. It takes the name of a view function and generates a URL that corresponds to that function.

```python
from flask import Flask
from flask import request

app = Flask(__name__)


@app.route('/')
def index():
    return 'This is the homepage'

@app.route('/about')
def about():
    return 'This is the about page'

@app.route('/contact')
def contact():
    return 'This is the contact page'


if __name__=="__main__":
   app.run(host="0.0.0.0") 

   with app.test_request_context():
        # generate a URL for the 'index' function
        index_url = url_for('index')
        print('URL for index:', index_url)

        # generate a URL for the 'about' function with a query parameter
        about_url = url_for('about', name='John')
        print('URL for about:', about_url)

        # generate a URL for the 'contact' function with a query parameter
        contact_url = url_for('contact', email='niyaj@example.com')
        print('URL for contact:', contact_url)
```

In this example, we have defined three view functions for the /, /about, and /contact routes. In the if __name__ == '__main__': block, we use the url_for() function to generate URLs for each of these routes.
